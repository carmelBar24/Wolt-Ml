```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def load_csv(file_name):
    df=pd.read_csv(file_name,encoding='latin-1')
    return df
```

```python
file_name="Data.csv"
df= load_csv(file_name)
Df_Restaurants=df.copy()
```

```python
Df_Restaurants=df.drop_duplicates(subset=["Name"],keep='first')
```

```python
Df_Rows_Num=Df_Restaurants.index
```

```python
print(Df_Rows_Num)
```

```python
for i in Df_Rows_Num:##שינוי עמודת הכסף לאורך הסטרינג כאשר כמות דולרים היא המספור
    Df_Restaurants["Money"][i]=(len(Df_Restaurants["Money"][i]))
    
```

```python
Df_Restaurants
```

```python
dict_cities={"ashdod":0,"beer-sheva":1,"hasharon":2,"herzliya":3,"jerusalem":4,"ness-ziona---rehovot":5,"netanya":6,"pardes-hanna":7,"petah-tikva":8,"rishon-lezion":9,"tel-aviv":10,"yavne":11}
```

```python
for key,value in dict_cities.items():
    Df_Restaurants["City"].replace({key:value}, inplace=True)
    
```

```python
Df_Restaurants["Delviry_Min"].replace({0:np.nan}, inplace=True)
Df_Restaurants["Delviry_Max"].replace({0:np.nan}, inplace=True)
```

```python
#df_res[col1].fillna(df_res[col1].median(),inplace=True)
for key,value in dict_cities.items():
    df_help=Df_Restaurants.loc[Df_Restaurants["City"]==value]
    median_min=df_help["Delviry_Min"].median()
    median_max=df_help["Delviry_Max"].median()
    print(median_min,median_max)
    Df_Restaurants.loc[(Df_Restaurants['City']==value)&(Df_Restaurants['Delviry_Min'].isnull()),'Delviry_Min']=median_min
    Df_Restaurants.loc[(Df_Restaurants['City']==value)&(Df_Restaurants['Delviry_Max'].isnull()),'Delviry_Max']=median_max
    
    
```
